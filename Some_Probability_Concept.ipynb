```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.stats import norm
```

```python
#generate a list with 500 element
a=[]
for i in range(500):
  a.append(np.random.random())

#500 samples from Bernoulli distribution
n=[]
V=0.25
def Prob(a,V):
  for i in range(len(a)):
    if(a[i]<=0.25):
      n.append(1)
    else:
      n.append(0)
Prob(a,V)
#probability mass calculation
empty_list = []
p=0.25
for i in n:
  e = p ** i * (1-p) ** (1-i)
  empty_list.append(e)
```

```python
#plot the probability mass functio using the selected data
plt.bar(n,empty_list,color=["Black" if val==0 else "blue" for val in n])
plt.title('Probability Mass Function')
plt.xlabel('Samples')
plt.ylabel('Probability')
plt.show()
```

```python
# Set the mean and standard deviation
mean = 100
std_dev = 5

# Generate 2000 samples from the normal distribution
samples = np.random.normal(mean, std_dev, 2000)

# Create a range of values for the PDF plot
x = np.linspace(80,120,1000)

# Calculate the PDF values for the range
pdf_values = norm.pdf(x, mean, std_dev)

# Plot the histogram
plt.hist(samples, bins=30,density='True',color='g', label='Histogram')

# Plot the PDF
plt.plot(x, pdf_values, 'r', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and PDF of Normal Distribution')

# Show the plot
plt.show()
```

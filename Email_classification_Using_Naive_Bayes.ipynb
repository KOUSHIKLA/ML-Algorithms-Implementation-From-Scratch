# **HOW TO CREATE DATA_FRAME**

```python
import numpy as np
import pandas as pd

# Number of attributes, classes, and data points
num_attributes = 3
num_classes = 3
num_data_points = 1000

# Initialize an empty DataFrame
df = pd.DataFrame()

# Generate data for each class
for i in range(num_classes):
    # Generate Multinomial distribution parameters
    params = np.random.dirichlet(np.ones(num_attributes),size=1)

    # Generate data points
    data = np.random.multinomial(1,params[0],size=int(num_data_points/num_classes))

    # Convert data points to DataFrame
    data_df = pd.DataFrame(data, columns=[f'Attribute_{j+1}' for j in range(num_attributes)])

    # Assign class
    data_df['Class'] = f'{i+1}'

    # Append to the main DataFrame
    df = df.append(data_df, ignore_index=True)

# Save the DataFrame to a CSV file
df.to_csv('classification_data.csv', index=False)
```

# **EMAIL CLASSIFICATION**

```python
# Training data
docs = ["send us your internet banking password",
        "send us your mobile number",
        "change your internet banking password",
        "change your userid",
        "send your internet banking password",
        "send us your account details"]
labels = ["spam", "ham", "ham", "spam", "spam", "spam"]

# Split the documents into words
words = [doc.split() for doc in docs]

# Create a set of all words
vocab = set([word for doc in words for word in doc])

# Count the occurrences of each word in each class
word_counts = {label: {word: 0 for word in vocab} for label in set(labels)}
for doc, label in zip(words, labels):
    for word in doc:
        word_counts[label][word] += 1

# Count the total number of documents in each class
doc_counts = {label: labels.count(label) for label in set(labels)}

# Calculate the prior probabilities of each class
priors = {label: count / len(labels) for label, count in doc_counts.items()}

# Calculate the likelihoods of each word in each class
likelihoods = {label: {word: (count + 1) / (sum(word_counts[label].values()) + len(vocab)) for word, count in counts.items()} for label, counts in word_counts.items()}

# Test data
test_docs = ["change your account number", "send your address"]

# Classify the test data
for test_doc in test_docs:
    # Split the document into words
    test_words = test_doc.split()

    # Calculate the posterior probabilities of each class
    posteriors = {}
    for label in set(labels):
        posteriors[label] = priors[label]
        for word in test_words:
            if word in likelihoods[label]:
                posteriors[label] *= likelihoods[label][word]

    # Predict the class with the highest posterior probability
    prediction = max(posteriors,key=posteriors.get)

    print(f'The email "{test_doc}" is classified as {prediction}.')
```
